```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/ToyotaCorolla_processed.csv")
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 5))
ax[0].plot(df["Price"], df["KM"], "bo")
ax[1].plot(df["Price"], df["Weight"], "bo")
ax[2].plot(df["Price"], df["HP"], "bo")
```

```python
fig, ax = plt.subplots(2, 2, figsize=(14, 10))
sns.boxplot(x=df["Automatic"], y=df["Price"], ax=ax[0][0])
sns.boxplot(x=df["Fuel_Type"], y=df["Price"], ax=ax[0][1])
sns.boxplot(x=df["Sport_Model"], y=df["Price"], ax=ax[1][0])
```

```python
sns.histplot(df['Price'], kde=True)
```

```python

fig, ax = plt.subplots(1, 3, figsize=(18, 5))
sns.countplot(x='Fuel_Type', data=df, ax=ax[0])
sns.countplot(x='Doors', data=df, ax=ax[1])
sns.countplot(x='ABS', data=df, ax=ax[2])
```
